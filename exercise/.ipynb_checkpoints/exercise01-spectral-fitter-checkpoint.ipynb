```python
#put here, as comments, a list of functions that you would like to have with possible function names

# read data
class read_data:
    def __init__(self, data_filename):
        self.data_filename = data_filename
        
    #readlines 
    def read_lines(self):
        data_dir = os.path.join(main_dir,"data")
        print("data dir is %s" % data_dir)
        divine_data_file = open(os.path.join(data_dir,data_filename),"r")
        #the readlines function gives a list with a line each
        divine_lines = divine_data_file.readlines()
        return divine_lines

    #create an output list with only the lines that we want
    def output_data(self, divine_lines):
        out_list = []
        
        for line in divine_lines:
            line = line.strip('\n')
            if len(line) > 0 and line[0] != '#':
                out_list.append(line)
        data_split = [line.split() for line in out_list]
        data_array = np.array(data_split, dtype=float)
        x = data_array[:, 0]
        y = data_array[:, 1]
        return x, y

    def read_fits(self):
        data_dir = os.path.join(main_dir,"data")
        print("data dir is %s" % data_dir)

#plot data 
class plot_data:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def plot(self):
        plt.plot(self.x, self.y, '.', label="data")
        plt.xlabel('velocity relative to LSR [km/s]')
        plt.ylabel('antenna temperature [K]')
        plt.show()

#Guassian function 
class fit_gaussian:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    #define gaussian function
    def gaussian(self, x, A, mu, sigma):
        return A * np.exp(-(x - mu)**2 / (2 * sigma**2))
    
    #set fitting range and mask here
    def fit(self, x_range=None):
        x_mask_range = (self.x >= x_range[0]) & (self.x <= x_range[1])
        x_fit = self.x[x_mask_range]
        y_fit = self.y[x_mask_range]


        initial_guess = [1.0, np.mean(x_fit), np.std(x_fit)]  # Initial guess for the fit parameters
        fit_params, covariance_matrix = curve_fit(self.gaussian, x_fit, y_fit, p0=initial_guess)
        gaussian_fit = self.gaussian(self.x, *fit_params)
        return gaussian_fit, fit_params, covariance_matrix

```

```python
#import the FITS module inside Astropy
from astropy.io import fits
from matplotlib.colors import LogNorm
%matplotlib inline
import glob
```

```python
#Class for FITS analysis
#plot fits data
class plot_fits:
    def __init__(self, filename):
        self.filename = filename

    def plot_spectrum(self):
        fits_image = fits.open(self.filename)
        pic = fits_image[0]
        data = pic.data[0, 0, :]

        plt.plot(data)
        plt.xlabel('')
        plt.ylabel('intensity')
        plt.title('spectrum')
        plt.show()
        
#Guassian function 
class fit_gaussian:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        
    #define gaussian function
    def gaussian(self, x, amplitude, mean, stddev):
        return amplitude * np.exp(-(x - mean)**2 / (2 * stddev**2))
    
    #set fitting range and mask here
    def fit(self, x_range=None):
        x_mask_range = (self.x >= x_range[0]) & (self.x <= x_range[1])
        x_fit = self.x[x_mask_range]
        y_fit = self.y[x_mask_range]

        initial_guess = [1.0, np.mean(x_fit), np.std(x_fit)]  #initial guess for fit parameters
        fit_params, covariance_matrix = curve_fit(self.gaussian, x_fit, y_fit, p0=initial_guess)
        gaussian_fit = self.gaussian(self.x, *fit_params)
        return gaussian_fit, fit_params, covariance_matrix
```

```python
#input file path
input_filename = os.path.join(data_dir, "salsa-vale-spectrum_50537.fits")

#plot spectrum
fits_plot = plot_fits(input_filename)
fits_plot.plot_spectrum()

out_name = os.path.join(results_dir,"data.fits.png")
plt.savefig(out_name)
```

```python
#load fits data
fits_data = fits.open(input_filename)
pic = fits_data[0]
data = pic.data[0, 0, :]

#create x, y data
x = np.arange(len(data))
y = data

#fitgaussian and set fitting range
FitGaussian = fit_gaussian(x, y)
x_range = [105, 140]
gaussian_fit, fit_params, covariance_matrix = FitGaussian.fit(x_range)

#plot data and gaussian fit
plt.plot(x, y, label='data')
plt.plot(x, gaussian_fit, label='Gaussian fit')
plt.xlabel('')
plt.ylabel('intensity')
plt.legend(loc="upper left")
plt.show()

#fitting result
print('fit_param[0]:', fit_params[0])
print('mean:', fit_params[1])
print('sigma:', fit_params[2])

out_name = os.path.join(results_dir,"GaussianFit.fits.png")
plt.savefig(out_name)
```
